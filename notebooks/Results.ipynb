# Results and Analysis

In this notebook we present the results achieved in the estimation of the ground state energy for different hamiltonians. In particular, four hamiltonians are analysed: two hamiltonians named as $H_3$ and $H_4$, where the subindex indicates the number of qubits, and two other hamiltonians corresponding to the hydrogen $H_{2}$ and water $H_{2}O$ molecules.

The ground state energy estimation is performed with different classical methodologies such as SPSA, L_BFGS_B, Bayesian Optimisation (BO) and other more novel ones such as Reinforcement Learning (RL).

```python
import matplotlib.pylab as plt
import numpy as np
import json
import sys
sys.path.insert(1, '../results/')
```

## Hamiltonian $H_{4}= X_{0}Z_{1}X_{2} + X_{0}Z_{2}X_{3} + Z_{0}X_{2}+ X_{1}Z_{3}$
Main features:
- Circuit requires optimising 16 parameters
- 4 qubits

```python
## H4
#SPSA
with open('../results/vqe/results_hamiltonian_4qubits_spsa.json', 'r') as f:
    results_spsa = json.load(f)

#L_BFGS_B
with open('../results/vqe/results_hamiltonian_4qubits_lbfgsb.json', 'r') as f:
    results_lbfgsb = json.load(f)

#BO Observables
with open('../results/vqe/results_hamiltonian_4qubits_bo_observables.json', 'r') as f:
    results_bo_observables = json.load(f)
#BO Statevector
with open('../results/vqe/results_hamiltonian_4qubits_bo_statevector.json', 'r') as f:
    results_bo_statevector = json.load(f)
ground_state_energy_H4 = results_spsa['real_ground_state_energy']
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,7), facecolor='white', constrained_layout = True, sharex=False, sharey=False)

plt.suptitle('Eigenvalues during optimisation')

n_samples = np.arange(len(results_spsa['history_best_eigenvalue_runs'][0]))
ax[0].set_title(r'$H_{4}$')

ax[0].plot(n_samples, np.mean(np.array(results_spsa['history_best_eigenvalue_runs']),0),'-o', label='SPSA', linewidth=1.3, markersize=2.5, color='C0')
ax[0].plot(n_samples, np.mean(np.array(results_lbfgsb['history_best_eigenvalue_runs']),0),'-o', label='L_BFGS_B', linewidth=1.3, markersize=2.5, color='C1')
ax[0].plot(n_samples, np.mean(np.array(results_bo_observables['history_best_eigenvalue_runs']),0),'-o', label='BO Observables', linewidth=1.3, markersize=2.5, color='C2')
ax[0].plot(n_samples, np.mean(np.array(results_bo_statevector['history_best_eigenvalue_runs']),0),'-o', label='BO Statevector', linewidth=1.3, markersize=2.5, color='C3')
ax[0].axhline(ground_state_energy_H4, linestyle='--',color='r', label='Ground state energy $E_{0}$')

ax[1].set_title('Eigenvalue - $E_{0}$')
ax[1].plot(n_samples, np.mean(np.array(results_spsa['history_best_eigenvalue_runs']),0) - ground_state_energy_H4,'-o', label='SPSA', linewidth=1.3, markersize=2.5, color='C0')
ax[1].plot(n_samples, np.mean(np.array(results_lbfgsb['history_best_eigenvalue_runs']),0) - ground_state_energy_H4,'-o', label='L_BFGS_B', linewidth=1.3, markersize=2.5, color='C1')
ax[1].plot(n_samples, np.mean(np.array(results_bo_observables['history_best_eigenvalue_runs']),0) - ground_state_energy_H4,'-o', label='BO Observables', linewidth=1.3, markersize=2.5, color='C2')
ax[1].plot(n_samples, np.mean(np.array(results_bo_statevector['history_best_eigenvalue_runs']),0) - ground_state_energy_H4,'-o', label='BO Statevector', linewidth=1.3, markersize=2.5, color='C3')


ax[1].set_yscale('log')
ax[0].set_ylabel('Eigenvalue')
ax[1].set_ylabel('Eigenvalue - $E_{0}$')
ax[1].set_xlabel('Number of iterations')
ax[0].legend(loc='upper right')
ax[1].legend(loc='upper right')

plt.legend()
plt.show()

```

## Analysis of $H_{3}$ and $H_{4}$
The convergence diagrams for Hamiltonians $H_{3}$ and $H_{4}$ reveal distinct optimisation patterns and outcomes. Both circuits associated with these Hamiltonians were optimised for up to 200 iterations, although some systems terminated earlier due to the configurations of the L-BFGS-B algorithm.
A clear disparity in results is evident between $H_3$ and $H_4$, with $H_3$ demonstrating superior performance. This significant difference can be attributed to several factors, including the number of circuit parameters, the intrinsic complexity of the Hamiltonians, and the scalability of the optimisation algorithms.

For $H_3$, the optimisation algorithms —particularly BO— yielded excellent results characterized by high convergence rates. This efficiency allowed for accurate estimates of ground state energy with relatively few iterations. In contrast, $H_4$ presented more substantial challenges; none of the algorithms achieved similarly favorable results within the same number of iterations. The difficulties encountered were largely due to an increased number of parameters, a higher qubit count, and greater overall system complexity. 

This analysis underscores the impact of system complexity and scale on the efficiency and effectiveness of quantum circuit optimisation techniques, highlighting the need for more robust or tailored approaches when addressing more complex Hamiltonians like $H_4$.

## Analysis of Hydrogen $H_{2}$ and Water $H_{2}O$


The ground state energy estimation graphs for hydrogen and water molecules reveal significant insights into the optimisation process for molecular systems of varying complexity. As the complexity of the system increases, so does the number of iterations required to achieve satisfactory results.

The models based on Bayesian Optimisation and Reinforcement Learning yielded particularly interesting outcomes. For the hydrogen molecule, the BO-based model demonstrated excellent optimisation in finding the ground state energy. However, when applied to the more complex $H_{2}O$ molecule, the results were considerably less impressive. This stark contrast in performance can be attributed to BO's limitations in handling high-dimensional optimisation spaces. The transition from optimising 16 parameters for $H_{2}$ to 56 parameters for $H_{2}O$ significantly impacted the algorithm's efficacy.

In contrast, the RL-based approach showed more consistent performance across both molecules. This robustness in the face of increased dimensionality is largely due to the use of neural networks in RL algorithms. Neural networks are inherently better equipped to handle high-dimensional spaces, allowing the RL model to maintain its effectiveness even as the complexity of the molecular system increased.

For more intricate systems, RL techniques become increasingly crucial in addressing the computational challenges posed by higher dimensionality. RL methods often scale more efficiently in high-dimensional settings and can handle larger datasets without the computational bottlenecks typically encountered by BO with Gaussian Processes (GP). This scalability makes RL particularly advantageous for optimising complex quantum systems, where the dimensionality of the problem can quickly overwhelm traditional BO approaches.

These findings highlight the importance of selecting appropriate optimisation techniques based on the quantum system's complexity. While BO may excel in simpler systems, advanced methods such as RL offer greater versatility and scalability for more complex molecular simulations. As we advance toward optimising increasingly sophisticated quantum systems, the capacity to efficiently handle high-dimensional spaces will be a critical factor in choosing the most effective optimisation algorithms.
